```python
import numpy as np
from matplotlib import pyplot as plt, cm

from src.pde import Pde, solve_heat_fdm, FdmMethod
```

```python
pde1 = Pde(
    kappa=1 / (np.pi**2),
    f=lambda x, t: 0,
    a=1,
    t_big=1,
    mu=lambda x: 1 / (np.pi**2) * np.sin(np.pi * x),
    mu1=lambda x: 0,
    mu2=lambda x: 0,
)

u1_i = solve_heat_fdm(pde1, 20, 30, FdmMethod.Implicit)
fig1, ax1 = plt.subplots(subplot_kw={"projection": "3d"})
x1, t1 = np.meshgrid(u1_i[0], u1_i[1])
surf1 = ax1.plot_surface(x1, t1, u1_i[2], cmap=cm.viridis, alpha=0.7)
ax1.view_init(elev=30, azim=45, roll=0)
plt.show()
```

```python
# n=17 -- fine, n=18 -- small, n=19 -- medium, n=20 -- large
u1_e = solve_heat_fdm(pde1, 20, 30, FdmMethod.Explicit)
fig2, ax2 = plt.subplots(subplot_kw={"projection": "3d"})
x2, t2 = np.meshgrid(u1_e[0], u1_e[1])
surf2 = ax2.plot_surface(x2, t2, u1_e[2], cmap=cm.plasma, alpha=0.7)
ax2.view_init(elev=30, azim=45, roll=0)
plt.show()
```

```python
pde2 = Pde(
    kappa=1 / (np.pi**2),
    f=lambda x, t: 0,
    a=1,
    t_big=0.1,
    mu=lambda x: 1 / (np.pi**2) * np.sin(np.pi * x),
    mu1=lambda x: 0,
    mu2=lambda x: 0,
)

fig = plt.figure()
ax = fig.add_subplot(projection="3d")

u2_i = solve_heat_fdm(pde2, 20, 20, FdmMethod.Implicit)
x1, t1 = np.meshgrid(u2_i[0], u2_i[1])
surf1 = ax.plot_surface(x1, t1, u2_i[2], cmap=cm.viridis, alpha=0.7)

u2_e = solve_heat_fdm(pde2, 20, 20, FdmMethod.Explicit)
x2, t2 = np.meshgrid(u2_e[0], u2_e[1])
surf2 = ax.plot_surface(x2, t2, u2_e[2], cmap=cm.plasma, alpha=0.7)

ax.view_init(elev=30, azim=45, roll=0)
plt.show()
```
