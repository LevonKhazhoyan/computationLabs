```python
import numpy as np
from matplotlib import pyplot as plt

from src.ode import Ode, solve_fdm
from math import sin, log

ode1 = Ode(
    p=lambda x: -(x - 2) / (x + 2),
    q=lambda x: x,
    r=lambda x: (sin(x)),
    f=lambda x: x**2,
    a=-1,
    b=1,
    alpha1=1,
    alpha2=0,
    alpha=0,
    beta1=1,
    beta2=0,
    beta=0,
)
u1 = solve_fdm(ode1, 2, 10e-9)
plt.plot(u1[0], u1[1])
plt.show()
```

```python
plt.plot(u1[2][0], u1[2][1])
plt.yscale("log")
plt.xscale("log")
plt.show()
```
